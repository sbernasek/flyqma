```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from clones.annotation.model_selection import ModelSelection
from growth.measure import LognormalSampler, MultiLognormalSampler
```

```python
generative_model = MultiLognormalSampler(ambiguity=0.2)
fig, ax = plt.subplots(figsize=(3, 2))
generative_model.show_pdf(ax=ax)
ax.set_xlim(0, 4)
ax.set_ylim(0, 1.1)
```

```python
sample_size = 1000
fig, ax = plt.subplots(figsize=(3, 2))
ind = np.repeat([0,1,2], repeats=sample_size)
sample = generative_model.sample(ind)
_ = ax.hist(sample, bins=100, alpha=0.3, density=True, color='k')
```

```python
selector = ModelSelection(sample, 'sample', min_num_components=1, max_num_components=5)
selector.plot_information_criteria()
```

```python
fig, ax = plt.subplots(figsize=(3, 2))
generative_model.show_pdf(ax=ax)
selector.BIC_optimal.plot_pdfs(ax=ax)
ax.set_xlim(0, 4)
ax.set_ylim(0, 1.5)
```
