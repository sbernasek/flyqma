# Load microscopy data

Confocal images are of a bleedthrough control experiment consisting of three color channels:
   - BLUE: Nuclear stain
   - RED: Clonal marker
   - GREEN: Control reporter (expressed equally within a vertical stripe of cells near the furrow)
   
We will use the first layer of the 4th stack as an example.

```python
stack_id = 4
layer_id = 0
```

```python
# load experiment
experiment = Experiment('../../../data/drosophila/')

# load image stack
stack = experiment.load_stack(experiment.stack_ids[stack_id], full=True)
layer = stack[layer_id]

# aggregate measurement data
measurements = stack.aggregate_measurements(raw=False)

# filter cells marked for exclusion
#measurements = measurements[measurements.selected]
```

```python
mixture_model = stack.annotator.classifier

from matplotlib.colors import ListedColormap
cmap = ListedColormap(['y', 'c', 'm'], N=3)
cmap.set_bad(alpha=0.)

mixture_model.set_cmap(cmap)

fig, ax = plt.subplots(figsize=(1, 1))
centroids = np.exp(mixture_model.model.means_[mixture_model.order])
ax.scatter(*centroids.T, s=25, c=mixture_model.label_colors[mixture_model.order])

#ax.set_xlim(0, xmax)
#ax.set_ylim(0, ymax)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_xlabel('Clonal marker level, X')
ax.set_ylabel('Local average, <X>')

savefig(fig, 'clustering')
```

```python
from matplotlib.colors import Normalize

cmap = plt.cm.Greys

fig = marginal.plot_cdfs(log=False, cmap=cmap, alpha=1.)
ax = fig.axes[0]

norm = Normalize(marginal.model.lbound, marginal.model.ubound)
order = np.argsort(marginal.means)
means = np.exp(marginal.means[order])
heights = 1.2*np.ones(marginal.num_components)
ax.scatter(means, heights, c=[cmap(norm(mu)) for mu in marginal.means[order]])
ax.set_xlabel('Clonal marker level')
ax.set_ylabel('Cumulative density')

savefig(fig, 'marginal_mixture_model_cdf')
```

# Community annotation

```python
from clones.annotation.spatial.sampling import CommunitySampler
community_sampler = CommunitySampler(layer.graph, 'r', depth=1.)
community_sampler.plot_autocorrelation()
fig = plt.gcf()
fig.axes[0].set_ylim(-0.05, 1.05)

#savefig(fig, 'sampling_autocorrelation_community')
```

```python
layer.graph.assign_community(community_sampler.level)

# add graph
graph_kw = dict(colorby='community', node_size=1., node_alpha=1., lw=0.5, edge_alpha=1., disconnect=True, node_color=None, edge_color=None)
#image_kw = image_kw=dict(cmap='r')
#fig = layer.plot_graph(figsize=figsize, graph_kw=graph_kw, image_kw=image_kw)

#fig, ax = plt.subplots(figsize=figsize)
#ax.axis('off')

fig = layer.get_channel('r').show(segments=False, cmap='r', figsize=figsize)
layer.graph.show(ax=fig.axes[0], **graph_kw, rasterized=True)

#savefig(fig, 'sampling_communities')
```

```python
marginal = stack.annotator.classifier[0]
layer.data['labels_marginal'] = marginal(layer.data)
kmask = layer.build_attribute_mask('labels_marginal')
fig = layer.get_channel('r').show(segments=False, cmap='r', figsize=figsize)
cmap = ListedColormap(['y', 'c', 'm'], N=3)
cmap.set_bad(alpha=0.)
fig.axes[0].imshow(kmask, cmap=cmap, vmin=0, vmax=2, alpha=1.)

#savefig(fig, 'classification_marginal')
```
