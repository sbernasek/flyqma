# Generate synthetic data

Load simulated bleedthrough control experiment and generate synthetic fluorescence data. Synthetic images contain three color channels:
   - BLUE: Nuclear stain
   - RED: Clonal marker
   - GREEN: Control reporter

```python
sweep = Sweep.load('../../../data/synthetic/sweeps/Sweep_181201_191516/')
simulation = sweep[0]
```

```python
# generate synthetic data

measurement_kwargs = dict(measurement_noise=0.)
microscopy_kwargs = dict(bg_level=0.2, bg_noise=0.3)

microscopy = simulation.generate_microscopy(ambiguity=0.2, rho=0.0, bleedthrough=0.5, 
                measurement_kwargs=measurement_kwargs, microscopy_kwargs=microscopy_kwargs)

# save synthetic data
#np.save('../../../data/synthetic/annotation_example/image.npy', microscopy.im)
#microscopy.data.to_hdf('../../../data/synthetic/annotation_example/data.hdf', key='microscopy')

# load existing synthetic data
microscopy.im = np.load('../../../data/synthetic/annotation_example/image.npy')
microscopy.data = pd.read_hdf('../../../data/synthetic/annotation_example/data.hdf', key='microscopy')
```

```python
# draw nuclei
from copy import deepcopy
microscopy.initialize()
microscopy.draw_clonal_marker()
means = microscopy.data['control'].values
stds = microscopy.data['control_std'].values
microscopy.draw_nuclei(channel=2, means=means, stds=stds)
gfp_raw = deepcopy(microscopy.im[2])
microscopy.add_bleedthrough(1, 2, microscopy.bleedthrough)
```

```python
clonal_marker = microscopy[1]
control = microscopy[2]
vmax = np.percentile(control.foreground_pixels, q=95)
fig0 = clonal_marker.render(size=2)
savefig(fig0, 'clonal_marker')

fig1 = control.render(size=2)
savefig(fig1, 'control_final')
```

```python
# evaluate bleed
bleed = clonal_marker.im[0]*microscopy.bleedthrough
gfp_background = (control.im[0] - gfp_raw) - bleed
gfp_original = gfp_background + gfp_raw

# plot original gfp signal
fig0, ax = plt.subplots(figsize=(2, 2))
ax.imshow(gfp_original.T, cmap=plt.cm.Greys, vmin=0, vmax=vmax)
ax.invert_yaxis()
ax.axis('off')

# plot rfp-gfp bleed
fig1, ax = plt.subplots(figsize=(2, 2))
ax.imshow(bleed.T, cmap=plt.cm.Greys, vmin=0, vmax=vmax)
ax.invert_yaxis()
ax.axis('off')

savefig(fig0, 'control_raw')
savefig(fig1, 'bleed')
```

```python
fig = microscopy.render(size=2)
savefig(fig, 'merge')
```
