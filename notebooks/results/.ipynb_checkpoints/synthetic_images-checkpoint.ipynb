```python
G = nx.read_pajek(filename)
communities = {}
for node in infomapWrapper.iterLeafNodes():
    communities[node.physicalId] = node.moduleIndex()

communities = infomapWrapper.getModules(level=1)
nx.set_node_attributes(G, values={str(k):v for k,v in communities.items()}, name='community')
drawNetwork(G)
```

```python
from matplotlib import colors

def drawNetwork(G):
    # position map
    pos = nx.kamada_kawai_layout(G)
    # community ids
    communities = [v for k,v in nx.get_node_attributes(G, 'community').items()]
    numCommunities = max(communities) + 1
    # color map from http://colorbrewer2.org/
    cmapLight = colors.ListedColormap(np.random.random((numCommunities, 3)), 'indexed', numCommunities)
    
    # Draw edges
    nx.draw_networkx_edges(G, pos)

    # Draw nodes
    nodeCollection = nx.draw_networkx_nodes(G,
        pos = pos,
        node_color = communities,
        cmap = cmapLight
    )
    # Set node border color to the darker shade
    darkColors = [cmapLight(v) for v in communities]
    nodeCollection.set_edgecolor(darkColors)

    # Draw node labels
    for n in G.nodes():
        plt.annotate(n,
            xy = pos[n],
            textcoords = 'offset points',
            horizontalalignment = 'center',
            verticalalignment = 'center',
            xytext = [0, 0],
            color = cmapLight(communities[int(n)-1])
        )

    plt.axis('off')
```
